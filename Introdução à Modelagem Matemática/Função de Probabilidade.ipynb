```python
import numpy as np

# Função de probabilidade
def Prob(M, i):
    P_dias = np.linalg.matrix_power(M, 6) @ i
    prob_Rico = P_dias[0]
    prob_ClasseMedia = P_dias[1]
    prob_Pobre = P_dias[2]
    prob_Devendo = P_dias[3]

    print("- Probabilidade de estar no estado Rico após 6 dias:", prob_Rico)
    print("- Probabilidade de estar no estado Classe média após 6 dias:", prob_ClasseMedia)
    print("- Probabilidade de estar no estado Pobre após 6 dias:", prob_Pobre)
    print("- Probabilidade de estar no estado Devendo após 6 dias:", prob_Devendo)
    print()

# Matriz de transição de Markov
M = [[0.05, 0.65, 0.20, 0.10],
    [0.05, 0.25, 0.30, 0.40],
    [0.10, 0.30, 0.35, 0.25],
    [0.05, 0.10, 0.30, 0.55]]

# Condições inciais
i_R = np.array([1, 0, 0, 0])
i_C = np.array([0, 1, 0, 0])
i_P = np.array([0, 0, 1, 0])
i_D = np.array([0, 0, 0, 1])

# Chamada da função para cada caso inicial
print("Probabilidade inicial de estar no estado Rico")
P_iR = Prob(M, i_R)
print("Probabilidade inicial de estar no estado Classe Média")
P_iC = Prob(M, i_C)
print("Probabilidade inicial de estar no estado Pobre")
P_iP = Prob(M, i_P)
print("Probabilidade inicial de estar no estado Devendo")
P_iD = Prob(M, i_D)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados da tabela
prices = np.array([0.59, 0.80, 0.95, 0.45, 0.79, 0.99, 0.90, 0.65, 0.79, 0.69, 0.79,
                   0.49, 1.09, 0.95, 0.79, 0.65, 0.45, 0.60, 0.89, 0.79, 0.99, 0.85])
sales_per_week = np.array([3980, 2200, 1850, 6100, 2100, 1700, 2000, 4200, 2440, 3300, 2300,
                           6000, 1190, 1960, 2760, 4330, 6960, 4160, 1990, 2860, 1920, 2160])

# Calcular as médias
mean_price = np.mean(prices)
mean_sales = np.mean(sales_per_week)

# Calcular c2 e c1
c2 = np.sum((prices - mean_price) * (sales_per_week - mean_sales)) / np.sum((prices - mean_price) ** 2)
c1 = mean_sales - c2 * mean_price

# Criar a curva de demanda ajustada
P_range = np.linspace(0.4, 1.1, 100)  # Intervalo de preços para plotagem
S_curve = c1 + c2 * P_range

# Plotar os dados e a curva de demanda ajustada
plt.scatter(prices, sales_per_week, label='Dados reais')
plt.plot(P_range, S_curve, color='red', label='Curva de demanda ajustada')
plt.xlabel('Preço (em dólares)')
plt.ylabel('Vendas por semana')
plt.legend()
plt.grid()
plt.title('Curva de demanda')
plt.show()

# Calcular o erro quadrático médio (RMSE)
predicted_sales = c1 + c2 * prices
rmse = np.sqrt(np.mean((sales_per_week - predicted_sales) ** 2))
print("Erro quadrático médio (RMSE):", rmse)
```

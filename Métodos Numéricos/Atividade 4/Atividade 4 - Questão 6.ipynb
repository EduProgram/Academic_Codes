```python
import numpy as np

def householder_reflection(v):
    v = np.array(v, dtype=float)
    v = v.reshape(-1, 1)
    v_norm = np.linalg.norm(v)
    v_hat = v.copy()
    v_hat[0] = v[0] + np.sign(v[0]) * v_norm
    v_hat_norm = np.linalg.norm(v_hat)
    if v_hat_norm == 0:
        return np.eye(len(v))
    else:
        return np.eye(len(v)) - 2 * np.dot(v_hat, v_hat.T) / np.dot(v_hat.T, v_hat)

def least_squares_householder(A, b):
    m, n = A.shape
    Q = np.eye(m)
    R = A.copy()

    for j in range(n):
        v = R[j:, j:j+1]
        H = np.eye(m)
        H[j:, j:] = householder_reflection(v)
        R = np.dot(H, R)
        if j < m - 1:
            Q = np.dot(H, Q)

    # Resolve o sistema triangular superior usando substituição para frente
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = (b[i] - np.dot(R[i, i+1:], x[i+1:])) / R[i, i]

    return x

# Exemplo de uso:
A = np.array([[2, 3], [-2, -6], [1, 0]])
b = np.array([3, -3, 6])
x = least_squares_householder(A, b)
print("Solução do sistema sobredeterminado: ", x)
```

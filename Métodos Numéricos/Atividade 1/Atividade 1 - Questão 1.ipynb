```python
import struct

def count_mantissa_bits(float_number):
    # Converte o número em ponto flutuante para sua representação binária em bytes
    binary = struct.pack('!d', float_number)

    # Extrai o expoente e a mantissa da representação binária
    bits = struct.unpack('!Q', binary)[0]
    mantissa_bits = bits & ((1 << 52) - 1)  # Máscara para extrair a mantissa

    # Conta o número de bits na mantissa
    count = sum(1 for _ in bin(mantissa_bits)[2:] if _ == '1')

    return count

print("Contagem de bits da mantissa em números de ponto flutuante")

while True:
    input_number = input("Insira um número em ponto flutuante (ou 's' para sair): ")

    if input_number.lower() == 's':
        break

    try:
        float_number = float(input_number)
        mantissa_bits = count_mantissa_bits(float_number)
        print(f"O número de bits da mantissa para {float_number} é: {mantissa_bits} bits")
    except ValueError:
        print("Entrada inválida. Tente novamente.")
```

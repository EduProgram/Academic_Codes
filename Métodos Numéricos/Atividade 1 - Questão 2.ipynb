```python
import math

def aproximacao_exp(x, n):
    aproximacao = 0.0
    exato = math.exp(x)
    for k in range(n+1):
        aproximacao += x**k / math.factorial(k)
    return aproximacao, exato

def calcular_erros(x_values, n):
    for x in x_values:
        aproximacao, exato = aproximacao_exp(x, n)
        erro_absoluto = abs(exato - aproximacao)
        erro_relativo = abs(exato - aproximacao) / exato
        print(f'x = {x}:')
        print(f'Aproximação: {aproximacao}')
        print(f'Exato: {exato}')
        print(f'Erro absoluto: {erro_absoluto}')
        print(f'Erro relativo: {erro_relativo}\n')

# Valores de x a serem avaliados
x_values = [-20, -10, -2, -1, 1, 2, 10, 20]

# Grau do polinômio da aproximação
n = 10

calcular_erros(x_values, n)
```

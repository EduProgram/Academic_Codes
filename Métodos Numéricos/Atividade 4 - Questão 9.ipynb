# Letra a)


```python
import numpy as np
from scipy.optimize import minimize

# Dados da tabela
t = np.array([1, 2, 3, 4, 5, 6, 7, 8])
c = np.array([8.0, 12.3, 15.5, 16.8, 17.1, 15.8, 15.2, 14.0])

# Função de erro a ser minimizada
def error_function(params):
    c1, c2 = params
    predicted_c = c1 * t * np.exp(c2 * t)
    return np.sum((predicted_c - c) ** 2)

# Chute inicial para os valores de c1 e c2
initial_guess = [1, 1]

# Minimizar a função de erro usando o método dos mínimos quadrados
result = minimize(error_function, initial_guess)

# Coeficientes c1 e c2 encontrados
c1_optimized, c2_optimized = result.x

print("c1 otimizado:", c1_optimized)
print("c2 otimizado:", c2_optimized)
```

# Letra b)

```python
# Calcular a concentração máxima
c_max = c1_optimized * (-1/c2_optimized) * np.exp(-1/c2_optimized)

# Calcular a meia-vida
half_life = np.log(2) / c2_optimized

print("Concentração máxima:", c_max)
print("Meia-vida:", half_life)
```

```python
import math

def f(x):
    return math.atan(x)

def f_prime(x):
    return 1 / (1 + x**2)

def newton_method(f, f_prime, x0, tolerance=1e-8, max_iterations=100):
    x = x0
    for i in range(max_iterations):
        fx = f(x)
        fx_prime = f_prime(x)

        if abs(fx_prime) < tolerance:
            print("Convergido: Derivada muito próxima de zero.")
            return x, i

        x_next = x - fx / fx_prime

        if abs(x_next - x) < tolerance:
            print("Convergido: Mudança em x muito pequena.")
            return x_next, i

        x = x_next

    print("Falhou em convergir dentro do número máximo de iterações.")
    return x, max_iterations

# Teste com aproximação inicial: x0 = 10
x0 = 10
root, iterations = newton_method(f, f_prime, x0)

print("Raiz aproximada:", root)
print("Número de iterações:", iterations)
print("Valor do arctan(raiz):", f(root))
```

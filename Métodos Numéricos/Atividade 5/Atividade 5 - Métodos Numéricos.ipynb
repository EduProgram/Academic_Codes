# Quest√£o 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
from scipy.interpolate import BarycentricInterpolator

# Dados da tabela
x_values = [1, 2, 5, 6, 7, 8, 10, 13, 17, 20, 23, 24, 25, 27, 27.7, 28, 29, 30]
f_values = [3.0, 3.7, 3.9, 4.2, 5.7, 6.6, 7.1, 6.7, 4.5, 7.0, 6.1, 5.6, 5.8, 5.2, 4.1, 4.3, 4.1, 3.0]

# Criar uma fun√ß√£o para interpola√ß√£o utilizando Lagrange
lagrange_interp = lagrange(x_values, f_values)

# Criar uma fun√ß√£o para interpola√ß√£o utilizando Newton
newton_interp = BarycentricInterpolator(x_values, f_values)

# Gerar pontos para o gr√°fico
x_range = np.linspace(min(x_values), max(x_values), 1000)
lagrange_y = lagrange_interp(x_range)
newton_y = newton_interp(x_range)

# Plotar os resultados
plt.figure(figsize=(10, 6))
plt.scatter(x_values, f_values, color='red', label='Pontos da Tabela')
plt.plot(x_range, lagrange_y, label='Interpola√ß√£o de Lagrange')
plt.plot(x_range, newton_y, label='Interpola√ß√£o de Newton')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Interpola√ß√£o Polinomial')
plt.legend()
plt.grid(True)
plt.show()
```

# Quest√£o 2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Dados da tabela
x_values = [1, 2, 5, 6, 7, 8, 10, 13, 17, 20, 23, 24, 25, 27, 27.7, 28, 29, 30]
f_values = [3.0, 3.7, 3.9, 4.2, 5.7, 6.6, 7.1, 6.7, 4.5, 7.0, 6.1, 5.6, 5.8, 5.2, 4.1, 4.3, 4.1, 3.0]

# Criar uma fun√ß√£o de spline c√∫bico
spline_interp = CubicSpline(x_values, f_values)

# Gerar pontos para o gr√°fico
x_range = np.linspace(min(x_values), max(x_values), 1000)
spline_y = spline_interp(x_range)

# Plotar os resultados
plt.figure(figsize=(10, 6))
plt.scatter(x_values, f_values, color='red', label='Pontos da Tabela')
plt.plot(x_range, spline_y, label='Spline C√∫bico')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Spline C√∫bico para Interpola√ß√£o')
plt.legend()
plt.grid(True)
plt.show()
```

```
import numpy as np
import pylab as pl
import sympy  as sp
import matplotlib.pyplot as plt

sp.init_printing()
```

```
  #Limites baseados no cilindro elíptico
  a,b = -2,2
  c,d = -2,2
  M= 8

#Eventos
e = [100,150,200,250,300,350,400,450,500, 1000,1500,10000,15000,100000,150000,1000000]

calc = np.empty(len(e)) # vetor par armarzenar volume calculado
pos = 0 #posição no vetor calc


for k in (e):
    v = 0 #Volume entre as superfícies

    for i in range(k):
      x = np.random.uniform(low = a, high = b)
      y = np.random.uniform(low = c, high = d)
      z = np.random.uniform(low = 0, high = M)
      if z < 8 - x**2 - y**2 and z > x**2 + 3*y**2 :
        v+=1.

    calc[pos] = (b-a)*(d-c)*(v/k)*M
    pos += 1
print(calc)
```

```
media = sum(calc) / float(len(calc))

dv = np.std(calc)

print('Média :', media)
print('Desvio Padrão: ',dv)


pl.hist(calc, bins=5)
pl.title('Distribuição normal')
pl.show()
```

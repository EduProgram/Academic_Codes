Em 1868, a introdução acidental nos Estados Unidos do inseto de algodão (Icerya purchasi) da Australia ameaçou destruir a indústria cítrica americana. Para remediar esta situacão, um predador australiano natural, besouro-joaninha (Novius cardinalis), foi importado daquele país. Os besouros mantiveram os insetos em um nível relativamente baixo. Quando o DDT (um inseticida) foi descoberto para matar insetos, os agricultores aplicaram-no na esperaça de reduzir ainda mais a população de insetos de algodão. No entanto, o DDT também se mostrou fatal para o besouro, e o efeito geral do uso do inseticida foi aumentar o número dos insetos de algodão. Seja que $C_n$ e $B_n$ os níveis de
população de insetos de algodão e besouros-joaninha, respectivamente, após n dias. O modelo pode ser escrito da forma:

$$ C_{n+1} = C_n +k_0C_n −k_1B_nC_n $$

$$ B_{n+1} = B_n −k_2B_n +k_3B_nC_n $$

onde $k_i$ são constantes positivas.

$(a)$ Quais suposições estão sendo implicitamente feitas sobre o crescimento de   cada espécie na ausencia das outras espécies?

$(b)$ Determine os pontos fixos do sistema predador-presa. Considere:

$$ (k_0, k_1, k_2, k_3) = (0.6,0.004,0.1,0.001) $$

$(c)$ Modifique o modelo para refletir um sistema predador-presa em que os agricultores aplicam (com regularidade) um inseticida que destrói tanto o predador do inseto quanto a presa do inseto a uma taxa proporcional aos números presentes. Escolha adequadamente os novos parâmetros do modelo, se assim for necessário.

$(d)$ Determine os pontos fixos do sistema predador-presa em que os agricultores aplicam inseticida.

Resolução da letra $(a)$
---
No caso da ausência de outras espécies, o crescimento se dá da seguinte forma:

**Insetos de algodão**: A população cresce indefinidamente, visto que o primeiro
termo na equação é constante, o segundo aumenta e o terceiro é zerado.

**Besouros-joaninha**: A população diminui indefinidamente, visto que o primeiro
termo na equação é constante, o segundo diminiu e o terceiro é zerado.

Algumas considerações a serem feitas:

*  Os valores de '$k$' são constantes positivas
*  Como os valores de '$k$' foram explicitados apenas na letra $(b)$ do exercício, considerou-se apenas o sinal das constantes

Resolução da letra $(b)$
---

Para determinar os pontos fixos do sistema predador-presa, será feita uma modelagem computacional, sendo o código encontrado abaixo:

```
# Importando a bibliotexa 'sympy' para resolver a equação simbolicamente 

import sympy as sp 

C, B = sp.var('C B')

C,B
```

```
# Declarando as constantes 'k'

k_0, k_1, k_2, k_3 = (0.6, 0.004, 0.1, 0.001) 

print (k_0, k_1, k_2, k_3)
```

```
# Definindo os lados da equação

lado_esquerdo = [C,B]

lado_direito = [C + k_0*C - k_1*B*C, B - k_2*B + k_3*B*C,]

# Representando e calculando o sistema das equações

A = [lado_esquerdo[i] - lado_direito[i] for i in range(len(lado_esquerdo))]

print(lado_esquerdo)
print(lado_direito)
print(A)
```

```
# Determinando os pontos fixos

sp.solve(A, (C,B), dict=True)

print (sp.solve(A, (C,B), dict=True))
```

---
De acordo com a modelagem feita, os pontos fixos do sistema predador-presa foram:

$$C = 0 \space ; B = 0$$
$$C = 100 \space ; B = 150$$

Como os valores igual a zero significam que a população não existe, essa
hipótese será descartada. Então os pontos fixos são: 

$$C = 100 \space ; B = 150$$

Com a introdução do inseticida, o sistema ficará da seguinte forma:

$$ C_{n+1} = C_n +k_0C_n −k_1B_nC_n - k_4C_n$$

$$ B_{n+1} = B_n −k_2B_n +k_3B_nC_n - k_5B_n$$

Sendo $k_4$ e $k_5$ constantes positivas relacionadas a taxa de mortalidade dos insetos de algodão e dos besouros-joaninha em contato com o inseticida, respectivamente.


Algumas considerações a serem feitas:




*  Cada inseto reage de maneira distinta ao insetida
*  O inseticida é aplicado na mesma regularidade de 'n' dias em que o modelo foi proposto
*   Os valores de '$k_4$' e '$k_5$' são constantes positivas

```
# Introduzindo um inseticida que mata a população 'C' e 'B'

import sympy as sp

CI, BI = sp.var('CI BI')

CI, BI
```

```
# Declarando mais constantes 'k'

k_4, k_5 = (0.002, 0.003)

print (k_4, k_5)
```

Como o modelo foi modificado, foi preciso definir novos parâmetros para o modelo, sendo eles representados por '$k_4$' e '$k_5$'. Então foi preciso definir adequadamente os seus valores. 

Abaixo estão algumas considerações sobre os valores adotados:

*   Os seus valores são menores que os demais '$k$', pois considerou-se que a influência do inseticida é menor que as interações entre a própria população e o outro inseto

*   Como o enunciado diz, o efeito geral do uso do inseticida é fazer com que a população de insetos de algodão aumente. Para que isso seja feito, o valor de de '$k_4$' adotado foi menor que o '$k_5$', pois isso faz com que o inseticida mate menos insetos de algodão do que besouros-joaninha

---

```
# Definindo os lados da equação

lado_esquerdo_I = [CI, BI]

lado_direito_I = [CI + k_0*CI - k_1*BI*CI - k_4*CI, BI - k_2*BI + k_3*BI*CI - k_5*BI,]

# Representando e calculando o sistema das equações

I = [lado_esquerdo_I[i] - lado_direito_I[i] for i in range(len(lado_esquerdo_I))]

print(lado_esquerdo_I)
print(lado_direito_I)
print(I)
```

```
# Determinando os pontos fixos

sp.solve(I, (CI,BI), dict=True)

print (sp.solve(I, (CI,BI), dict=True))
```

```
# Armazenando os valores das funções

import numpy as np
import pylab as pl

def evolui_sistema(x_0, k, n=50):
  k1,k2,k3,k4=k
  x=np.zeros((n,2))
  x[0,:]=x_0
  for i in range(0,n-1):
    x[i+1,0] = x[i,0] + k_0*x[i,0] - k_1*x[i,1]*x[i,0]
    x[i+1,1] = x[i,1] - k_2*x[i,1] + k_3*x[i,1]*x[i,0]

  return (x)       
```

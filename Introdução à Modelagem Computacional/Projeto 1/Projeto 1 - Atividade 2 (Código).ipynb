Para iniciar nosso trabalho vamos transcrever o que foi pedido no enunciado da atividade 2:

Um economista esta interessado na variacão do preço de um unico produto denominado. Observa-se
que um precço alto para o produto no mercado atrai mais fornecedores. Contudo, aumentar a quantidade do produto fornecido tende a baixar o preço. Sobre tempo, ha uma interação entre precço e oferta.O economista propos o seguinte modelo, em que Pn representa o preço do produto no mês n e Qn
representam a quantidade. Encontre os valores de equilíbrio para este sistema


  Pn+1 = Pn −0.1(Qn −500)


 Qn+1 = Qn +0.2(Pn −100)



```
import numpy as np
import matplotlib.pyplot as plt
import math as mt
import sympy as sp
```

```
#Vamos escrever a função para cálculos e plotagem dos gráficos de acordo com os casos solicitados.
def calcula (P_0, Q_0, m):

  n = np.zeros(m)
  n[0] = 0
  Q = np.zeros(m)
  Q[0] = Q_0
  P = np.zeros(m)
  P[0] = P_0

  for i in range(0,m-1):
    Q[i+1] = Q[i] + 0.2*(P[i]-100)
    P[i+1] = P[i] - 0.1*(Q[i]-500)
    n[i+1] = i+1

  plt.figure(figsize=(8,6))
  plt.plot(n, P, "rx-", label = '$P_n$')
  plt.plot(n, Q, "ko-", label="$Q_n$")
  plt.legend()
  plt.grid()
  plt.xlabel("Tempo (meses)")
  plt.title('Gráfico de Preço e Quantidade x Tempo')
  plt.show()

```

Vamos calcular os casos solicitados na atividade

```
# Caso A
calcula (100, 500, 50)

# Caso B
calcula (200, 500, 50)

# Caso C
calcula (100, 600, 50)

# Caso D
calcula (100, 400, 50)

```

```
#Vamos encontrar os valores de equilíbrio para cada caso, sendo os casos representados pelas letras A, B, C e D
casos = ['A','B','C', 'D']
Q_0 = [500, 500, 600, 400]
P_0 = [100, 200, 100, 100]

P, Q = sp.var ('P Q')

for i in range(0,len(Q_0)):
  lado_esq = [P, Q]
  lado_dir = [(P_0[i]-0.1*(Q_0[i]-500)), (Q_0[i] + 0.2*(P_0[i]-100))]

  A = [-lado_esq[j] + lado_dir[j] for j in range(len(lado_esq))]


  print('Os pontos de equilíbrio para o caso '+casos[i]+' são: ',sp.solve(A,(P,Q)))

```
